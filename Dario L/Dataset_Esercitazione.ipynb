```python
import pandas as pd
```

```python
#link: https://www.kaggle.com/saadalsharif/smartphone
mobile = pd.read_csv("../../mobile.csv")
mobile
```

```python
mobile ["name"]
```

```python
#solo modelli samsung
samsung = mobile [mobile["item_brand_name"]=="samsung"]
samsung
```

```python
#solo modelli apple
apple = mobile [mobile["item_brand_name"]=="apple"]
apple
```

```python
#solo modelli huawei
huawei = mobile [mobile["item_brand_name"]=="huawei"]
huawei
```

```python
#solo Lte 4g
LTE_4 = mobile [mobile["Generation"]== "4G LTE"]
LTE_4
```

```python
#tutto ma non Lte 4g
nonLTE_4 = mobile [mobile["Generation"]!= "4G LTE"]
nonLTE_4
```

```python
# modelli apple
modelli_apple = apple ["name"].unique()
modelli_apple.sort()
print(modelli_apple)
count = sum (1 for item in modelli_apple)
print ("\nNumero modelli tot: ",count)
```

```python
# modelli samsung
modelli_samsung = samsung ["name"].unique()
modelli_samsung.sort()
print(modelli_samsung)
count_samsung = sum (1 for item in modelli_samsung)
print ("\nNumero modelli tot: ",count_samsung)
```

```python
# modelli huawei
modelli_huawei = huawei ["name"].unique()
modelli_huawei.sort()
print(modelli_huawei)
count_huawei = sum (1 for item in modelli_huawei)
print ("\nNumero modelli tot: ",count_huawei)
```
