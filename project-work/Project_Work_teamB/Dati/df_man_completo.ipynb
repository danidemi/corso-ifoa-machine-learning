```python
import json
import math
import sys

def extract_averge_review(reviews):
    try:
        if (type(reviews) == float):
            return None
        
        if (type(reviews) != str):
            #print( f"WARN not a string but a {type(reviews)} {reviews}" )
            return None
        
        reviews = reviews.replace("\\\\", "\\")
        kvs = json.loads(reviews)
        ratings = []
        for kv in kvs:
            if("rating" in kv):
                #print( f"\n\n{kv}" )
                ratings.append( float( kv["rating"] ) )
                
        #print( f"\n\n{ratings}" )
        
        # questo blocco Ã¨ per fare qualcosa di utile con i voti di un prodotto.
        # ad esempio in questo caso viene calcolata la media delle recensioni
        if len(ratings) == 0:
            return None
        else:
            total = 0.0
            for r in ratings:
                total = total + r
            avg = total / len(ratings)
            return avg        
                
    except:
        err_type, err_value, err_traceback = sys.exc_info()
        #print( f"WARN '{err_type}' '{err_value}' parsing {reviews}" )
        return None


def extract_material(features_json):
    # [{"key":"Material","value":["Suede"]},{"key":"Gender","value":["Men"]},{"key":"Color","value":["Beige"]},{"key":"Model","value":["M254TNN"]},{"key":"Manufacturer Part Number","value":["M254TNN"]},{"key":"Brand","value":["New Balance Numeric"]},{"key":"Age Group","value":["Adult"]}]    
    try:
        if (type(features_json) == float):
            return None
        
        if (type(features_json) != str):
            print( f"WARN not a string but a {type(features_json)} {features_json}" )
            return None
        
        features_json = features_json.replace("\\\\", "\\")
        kvs = json.loads(features_json)
        for kv in kvs:
            if(kv["key"] == "Material"):
                return kv["value"][0]
            
        return None
    except:
        err_type, err_value, err_traceback = sys.exc_info()
        print( f"WARN '{err_type}' '{err_value}' parsing {features_json}" )
        return None
    
def extract_age_group(features_json):
    try:
        if (type(features_json) == float):
            return None
        
        if (type(features_json) != str):
            print( f"WARN not a string but a {type(features_json)} {features_json}" )
            return None
        
        features_json = features_json.replace("\\\\", "\\")
        kvs = json.loads(features_json)
        for kv in kvs:
            if(kv["key"] == "Age Group"):
                return kv["value"][0]
        return None
    except:
        err_type, err_value, err_traceback = sys.exc_info()
        print( f"WARN '{err_type}' '{err_value}' parsing {features_json}" )
        return None    

    

    
    
     

df_man_ridotto = pd.DataFrame()

df_man_ridotto["brand"] = man_shoes["brand"]
df_man_ridotto["categorie"] = man_shoes["categories"]
df_man_ridotto["averge_review"] = man_shoes["reviews"].apply( extract_averge_review )
df_man_ridotto["prices_amountMin"] = pd.to_numeric(man_shoes["prices.amountMin"], errors='coerce')
df_man_ridotto["prices_amountMax"] = pd.to_numeric(man_shoes["prices.amountMax"], errors='coerce')
df_man_ridotto["prices.currency"] = man_shoes["prices.currency"]
df_man_ridotto["colore"] = man_shoes["colors"].astype(str)
df_man_ridotto["size"] = man_shoes["sizes"]   
df_man_ridotto["age_group"] = man_shoes["features"].apply( extract_age_group )
df_man_ridotto["material"] = man_shoes["features"].apply( extract_material )


df_man_ridotto 
```

```python
#filtro materiale


sintetico = ['Plastic','Synthetics', 'Polyamide, Polyester, Polyurethane and Spandex', 'Polyester', 'Nylon', 'Acrylic', 'Synthetic', 'Synthetics', 'SS304 Stainless Steel / Nylon Coating', 'Stainless Steel']           
            
mesh = ['mesh', 'Mesh']
    
non_sintetico = ['Cotton blend', 'Textile', '100 percent cotton', 'Canvas', 'Fabric', 'Wool']     
  
misto = ['Denim', 'Leather and Mesh', 'leather-and-fabric', 'Down-filled', 'LEATHER']    
    
pelle = ['Leather', 'Suede', 'suede']    
    
def mat_sneakers(x):
    for i in x:
        if x in sintetico:
            return ("Sintetico")
        elif x in non_sintetico:
            return ('Non sintetico')
        elif x in misto:
            return ('Misto')
        elif x in pelle:
            return ("Pelle")
        elif x in mesh:
            return ('mesh')
```

```python

#filtro colori




to_save_col = ['Black', 'White','Blue','Red', 'Brown', 'Grey', 'Gray', 'Green', 'Yellow', 'Silver', "Dark",  "Fume",  "Camel", "Wheat", "Indigo", "Chocolate", 'Tan', 'Bone', 'Cigar', 'Niagra Falls',  'Orange', 'Burgundy Pull Up', 'Ball', 'Multi-color']



def col_sneakers(x):
    for col_to_save in to_save_col:
        if col_to_save in x : 
            return (col_to_save)
        
        
#filtro colori raggruppati        
        
blue  = ['Blue', 'Indigo']
black = ['Black', "Dark"]
brown = ['Brown', "Chocolate", 'Cigar',  "Camel", 'Tan']
white = ['White', 'Bone']
grey = ['Gray', 'Grey']

def cat_color_reduce (x):
    for i in x:
        if x in blue:
            return ('blue')
        elif x in black:
            return ('black')
        elif x in brown:
            return ('brown')
        elif x in white:
            return ('white')
        elif x in grey:
            return ('grey')
        else:
            return (x)        
```
