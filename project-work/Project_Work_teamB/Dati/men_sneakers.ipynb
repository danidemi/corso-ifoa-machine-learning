```python
import pandas as pd
man_shoes = pd.read_csv('Datafiniti_Mens_Shoe_Prices.csv', error_bad_lines=False)
man_shoes
```

```python
rows = man_shoes["categories"].unique()

s = set ()



for row in rows:
    #print("-- " + row)
    categories = row.split(",")
    for category in categories:
        ##print("---- " + category)
        s.add(category)
           

print(s)
```

```python
    
to_save = ["Driving Shoes", 'Running', "Men's Jackets & Outerwear",'Boat Shoes',
               'Family Recreation', 'Prada Sport','Sports & Fitness',  'Hiking', 'Cycling', 'Other Slip-Ons',  
               "Men's Sneakers & Athletic Shoes", 'Fitness & Cross-Training', 'Basketball',  'Athletic Socks',  
               'Walking Shoes', 'Racquet Sports', 'Running Shoes', 'Fashion Sneakers', 
               'Camping & Hiking', 'sports', 'Slip-ons',  'Skateboarding', 'Sneakers & Athletic Shoes', 
               'Team Sports', 'Sports', 'Sports & Outdoors', ' Skate & Street', 
               'Athletic', 'Outdoor', 'Outdoor Activities',  'Active & Performance', 'Active', 'Tennis',
               'Fitness & Sports', 'Exercise & Fitness', 'Hiking & Trekking', "Men's Casual Shoes"]

def cat_def_sneakers(x):
    for cat_to_save in to_save:
        if cat_to_save in x : 
            return (cat_to_save)
```

```python
man_shoes["macro_categoria"] = man_shoes["categories"].apply( cat_def_sneakers )

man_shoes
```
