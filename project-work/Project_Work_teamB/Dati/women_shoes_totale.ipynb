```python

import json
import math
import sys


def extract_material(features_json):
    # [{"key":"Material","value":["Suede"]},{"key":"Gender","value":["Men"]},{"key":"Color","value":["Beige"]},{"key":"Model","value":["M254TNN"]},{"key":"Manufacturer Part Number","value":["M254TNN"]},{"key":"Brand","value":["New Balance Numeric"]},{"key":"Age Group","value":["Adult"]}]    
    try:
        if (type(features_json) == float):
            return None
        
        if (type(features_json) != str):
            print( f"WARN not a string but a {type(features_json)} {features_json}" )
            return None
        
        features_json = features_json.replace("\\\\", "\\")
        kvs = json.loads(features_json)
        for kv in kvs:
            if(kv["key"] == "Material"):
                return kv["value"][0]
        return None
    except:
        err_type, err_value, err_traceback = sys.exc_info()
        print( f"WARN '{err_type}' '{err_value}' parsing {features_json}" )
        return None
```

```python
def extract_age_group(features_json):
    try:
        if (type(features_json) == float):
            return None
        
        if (type(features_json) != str):
            print( f"WARN not a string but a {type(features_json)} {features_json}" )
            return None
        
        features_json = features_json.replace("\\\\", "\\")
        kvs = json.loads(features_json)
        for kv in kvs:
            if(kv["key"] == "Age Group"):
                return kv["value"][0]
        return None
    except:
        err_type, err_value, err_traceback = sys.exc_info()
        print( f"WARN '{err_type}' '{err_value}' parsing {features_json}" )
        return None 
```

```python
def extract_reviews_rating(revdefiews_json):
    try:
        if (type(reviews_json) == float):
            return None
        
        if (type(reviews_json) != str):
            print( f"WARN not a string but a {type(reviews_json)} {reviews_json}" )
            return None
        
        reviews_json = reviews_json.replace("\\\\", "\\")
        kvs = json.loads(reviews_json)
        for kv in kvs:
            if(kv["key"] == "rating"):
                return kv["value"][0]
        return None
    except:
        err_type, err_value, err_traceback = sys.exc_info()
        print( f"WARN '{err_type}' '{err_value}' parsing {reviews_json}" )
        return None 
```

```python
reduced_women_shoes = pd.DataFrame()
reduced_women_shoes['brand'] = women_shoes['brand']
reduced_women_shoes['material'] = women_shoes['features'].apply(extract_material)
reduced_women_shoes['group_aged']= women_shoes['features'].apply(extract_age_group)
#redueced_women_shoes['rating_review'] = women_shoes['reviews'].apply( extract_reviews_rating)
reduced_women_shoes["prices.amountMin"] = pd.to_numeric(women_shoes["prices.amountMin"], errors='coerce')
reduced_women_shoes["prices.amountMax"] = pd.to_numeric(women_shoes["prices.amountMax"], errors='coerce')
reduced_women_shoes = reduced_women_shoes.dropna()
reduced_women_shoes
```
