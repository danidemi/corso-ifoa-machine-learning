```python
import pandas as pd
import numpy as np
import json
import math
import sys
```

```python
Woman_7210 = pd.read_csv('7210_1.csv', error_bad_lines=False, warn_bad_lines=True, low_memory=False)
Woman_7210
```

```python
def extract_color(features_json):
    # [{"key":"Material","value":["Suede"]},{"key":"Gender","value":["Men"]},{"key":"Color","value":["Beige"]},{"key":"Model","value":["M254TNN"]},{"key":"Manufacturer Part Number","value":["M254TNN"]},{"key":"Brand","value":["New Balance Numeric"]},{"key":"Age Group","value":["Adult"]}]    
    try:
        if (type(features_json) == float):
            return None
        
        if (type(features_json) != str):
            #print( f"WARN not a string but a {type(features_json)} {features_json}" )
            return None
        
        features_json = features_json.replace("\\\\", "\\")
        kvs = json.loads(features_json)
        for kv in kvs:
            if(kv["key"] == "Color")or (kv["key"] == "Colour"):
                return kv["value"][0]
            
        return None
    except:
        err_type, err_value, err_traceback = sys.exc_info()
        #print( f"WARN '{err_type}' '{err_value}' parsing {features_json}" )
        return None
    
Woman_7210['Colour1']= Woman_7210['features'].apply(extract_color)
```

```python
def extract_material(features_json):
    # [{"key":"Material","value":["Suede"]},{"key":"Gender","value":["Men"]},{"key":"Color","value":["Beige"]},{"key":"Model","value":["M254TNN"]},{"key":"Manufacturer Part Number","value":["M254TNN"]},{"key":"Brand","value":["New Balance Numeric"]},{"key":"Age Group","value":["Adult"]}]    
    try:
        if (type(features_json) == float):
            return None
        
        if (type(features_json) != str):
            print( f"WARN not a string but a {type(features_json)} {features_json}" )
            return None
        
        features_json = features_json.replace("\\\\", "\\")
        kvs = json.loads(features_json)
        for kv in kvs:
            if(kv["key"] == "Material"):
                return kv["value"][0]
            
        return None
    except:
        err_type, err_value, err_traceback = sys.exc_info()
        print( f"WARN '{err_type}' '{err_value}' parsing {features_json}" )
        return None

Woman_7210["material"] = Woman_7210["features"].apply( extract_material )
```

```python
Woman_7210=Woman_7210[['brand','categories', "material", "Colour1", "colors", 'prices.amountMin', 'reviews']]
```

```python
def v(x):
    if "Sneakers" in x:
        return "Sneakers"
    
Woman_7210['Sneakers']=Woman_7210['categories'].apply(v)
Woman_7210=Woman_7210[Woman_7210['Sneakers'].isnull()==False]
```

```python
def extract_averge_review(reviews):
    try:
        if (type(reviews) == float):
            return None
        
        if (type(reviews) != str):
            #print( f"WARN not a string but a {type(reviews)} {reviews}" )
            return None
        
        reviews = reviews.replace("\\\\", "\\")
        kvs = json.loads(reviews)
        ratings = []
        for kv in kvs:
            if("rating" in kv):
                #print( f"\n\n{kv}" )
                ratings.append( float( kv["rating"] ) )
        if len(ratings) == 0:
            return None
        else:
            return sorted(ratings)      
                
    except:
        err_type, err_value, err_traceback = sys.exc_info()
        #print( f"WARN '{err_type}' '{err_value}' parsing {reviews}" )
        return None 
    
Woman_7210["All_rating"] = Woman_7210["reviews"].apply( extract_averge_review )
Woman_7210=Woman_7210[ Woman_7210["reviews"].isnull()==False ].sort_values(by=['reviews'], ascending=False)
```

```python
def extract_text(reviews):
    try:
        if (type(reviews) == float):
            return None
        
        if (type(reviews) != str):
            #print( f"WARN not a string but a {type(reviews)} {reviews}" )
            return None
        
        reviews = reviews.replace("\\\\", "\\")
        kvs = json.loads(reviews)
        text = []
        for kv in kvs:
            if("text" in kv):
                #print( f"\n\n{kv}" )
                text.append( str( kv["text"] ) )
        if len(text) == 0:
            return None
        else:
            return text      
                
    except:
        err_type, err_value, err_traceback = sys.exc_info()
        #print( f"WARN '{err_type}' '{err_value}' parsing {reviews}" )
        return None 
    
Woman_7210["text_review"] = Woman_7210["reviews"].apply( extract_text )
```
