```python
import pandas as pd
```

```python
Woman_7210 = pd.read_csv('7210_1.csv', error_bad_lines=False, warn_bad_lines=True, low_memory=False)
Woman_7210
```

```python
Woman_7210=Woman_7210[Woman_7210['prices.currency']=='USD']
```

```python
Woman_7210['medium.price']=(Woman_7210['prices.amountMin']+Woman_7210['prices.amountMax'])/2
```

```python
Woman_7210=Woman_7210[['brand', 'categories','features','prices.amountMin','prices.amountMax','medium.price']]
Woman_7210
```

```python
rows = Woman_7210["categories"].unique()
#print(rows)
s = set()
for row in rows:
    #print("-- " + row)
    categories = row.split(",")
    for category in categories:
        ##print("---- " + category)
        s.add(category)

print(s)
```

```python
cat_def={"Womens' Athletic Shoes", 'Sneakers', "Women's Athletic Shoes", 'Fashion Sneakers', 'Outdoor', 'Sneakers & Athletic Shoes', 'Running'}
```

```python
def y(x):
    cat_def_woman_7210={"Womens' Athletic Shoes", 'Sneakers', "Women's Athletic Shoes", 'Fashion Sneakers', 'Outdoor', 'Sneakers & Athletic Shoes', 'Running'}
    p=''
    for i in cat_def_woman_7210:
        if i in x:
            p=p+i+', '
    return p[:-2]

def g(x):
    for i in x:
        return y(x)
    
Woman_7210['Categorie_DEF']= Woman_7210['categories'].apply(g)
Woman_7210=Woman_7210[Woman_7210['Categorie_DEF']!='']

Woman_7210
```

```python
import json
import math
import sys


def extract_material(features_json):
    # [{"key":"Material","value":["Suede"]},{"key":"Gender","value":["Men"]},{"key":"Color","value":["Beige"]},{"key":"Model","value":["M254TNN"]},{"key":"Manufacturer Part Number","value":["M254TNN"]},{"key":"Brand","value":["New Balance Numeric"]},{"key":"Age Group","value":["Adult"]}]    
    try:
        if (type(features_json) == float):
            return None
        
        if (type(features_json) != str):
            print( f"WARN not a string but a {type(features_json)} {features_json}" )
            return None
        
        features_json = features_json.replace("\\\\", "\\")
        kvs = json.loads(features_json)
        for kv in kvs:
            if(kv["key"] == "Material"):
                return kv["value"][0]
            
        return None
    except:
        err_type, err_value, err_traceback = sys.exc_info()
        print( f"WARN '{err_type}' '{err_value}' parsing {features_json}" )
        return None
    
Woman_7210_reduced = pd.DataFrame()
Woman_7210_reduced["brand"] = Woman_7210["brand"]
Woman_7210_reduced["material"] = Woman_7210["features"].apply( extract_material )
Woman_7210_reduced["Categorie_DEF"] = Woman_7210["Categorie_DEF"]
Woman_7210_reduced["prices.amountMin"] = Woman_7210["prices.amountMin"]
Woman_7210_reduced["prices.amountMax"] = Woman_7210["prices.amountMax"]
Woman_7210_reduced["medium.price"] = Woman_7210["medium.price"]

Woman_7210_reduced
```
